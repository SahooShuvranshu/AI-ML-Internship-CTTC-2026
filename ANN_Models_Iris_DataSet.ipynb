# Iris Data Set | Categorical

26th June 2026

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
iris = pd.read_csv(r"/content/iris.data", header=None)
iris
```

```python
iris.columns = ['SL', 'SW', 'PL', 'PW', 'FlowerType']
iris
```

```python
iris.isnull().sum()
```

```python
iris.dtypes
```

# Encoding

```python
from sklearn.preprocessing import LabelEncoder
lb = LabelEncoder()
```

```python
iris.FlowerType = lb.fit_transform(iris.FlowerType)
iris.FlowerType.unique()
```

```python
lb.inverse_transform([0,1,2])
```

# IP/OP Creation

```python
ip = iris.drop('FlowerType', axis=1)
op = iris.FlowerType
```

# To Categorical

```python
from keras.utils import to_categorical
op = to_categorical(op, 3)
```

```python
op
```

# Train Test Split

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(ip, op, test_size = 0.2)
```

# Standard Scaler

```python
from keras.models import Sequential
from keras.layers import Dense
from keras.optimizers import Adam
```

# Model Training

```python
model = Sequential()
model.add(Dense(20, input_dim=4, activation='relu'))

model.add(Dense(35, activation='relu'))
model.add(Dense(15, activation='relu'))

model.add(Dense(3, activation='softmax'))
model.compile(Adam(learning_rate=0.01), loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
model.summary()
```

```python
model.fit(x_train, y_train, validation_data=(x_test, y_test), epochs=15)
```

```python
pred = model.predict(x_test)
pred
```

```python
p = np.argmax(pred, axis=1)
p
```

```python
y_test
```

```python
y = np.argmax(y_test, axis=1)
y
```

# Accuracy, Recall, Precision, Average Calculations

```python
from sklearn.metrics import accuracy_score, recall_score, precision_score, f1_score

ac = accuracy_score(y, p)
rc = recall_score(y, p, average='micro')
pc = precision_score(y, p, average='micro')
f1 = f1_score(y, p, average='micro')

print('Accuracy Score: ', ac)
print('Recall Score: ', rc)
print('Precision Score: ', pc)
print('F1 Score: ', f1)
```
